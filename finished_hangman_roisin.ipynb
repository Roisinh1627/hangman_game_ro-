```python
# 1. Word bank (150 words, no repeats)
word_bank = [
    "planet", "castle", "bridge", "forest", "silver", "hunter", "puzzle", "dragon","window", "garden", "throne", "candle", "butter", "marble", "saddle", "rocket",
    "pirate", "anchor", "helmet", "shield", "wizard", "market", "fabric", "pencil","camera", "ticket", "branch", "stream", "flower", "animal", "desert", "island",
    "throat", "pocket", "button", "string", "circle", "square", "danger", "singer","writer", "reader", "doctor", "farmer", "bottle", "coffee", "cookie", "donkey",
    "rabbit", "kitten", "puppy", "parrot", "eagle", "falcon", "turtle", "shark","salmon", "pepper", "tomato", "cheese", "butter", "pickle", "banana", "orange",
    "grapes", "peach", "apples", "carrot", "potato", "onions", "garlic", "ginger","pepper", "cactus", "planet", "meteor", "galaxy", "cosmic", "rocket", "comet",
    "sunset", "sunrise", "thunder", "lighten", "stormy", "breeze", "winter", "summer","autumn", "spring", "harbor", "vessel", "bridge", "castle", "temple",
    "market","fabric", "cotton", "velvet", "leather", "rubber", "plastic", "copper", "silver","golden", "bronze", "marble", "granite", "crystal", "quartz", "jasper", 
    "onyx","velvet", "thread", "needle", "scarf", "jacket", "trousers", "sweater", "blanket","pillow", "basket", "ladder", "hammer", "chisel", "spirit", "memory", "future",
    "moment", "shadow", "bright", "silent", "noodle", "puzzle", "random", "simple","gentle", "strong", "broken", "smooth", "sticky", "crispy", "salmon", "butter"
]

HANGMAN_PICS = [
    """
       +---+
       |   |
           |
           |
           |
           |
    =========
    """,
    """
       +---+
       |   |
       O   |
           |
           |
           |
    =========
    """,
    """
       +---+
       |   |
       O   |
       |   |
           |
           |
    =========
    """,
    """
       +---+
       |   |
       O   |
      /|   |
           |
           |
    =========
    """,
    """
       +---+
       |   |
       O   |
      /|\  |
           |
           |
    =========
    """,
    """
       +---+
       |   |
       O   |
      /|\  |
      /    |
           |
    =========
    """,
    """
       +---+
       |   |
       O   |
      /|\  |
      / \  |
           |
    =========
    """
]

# 2. IMPORTS
import random

# Pluralisation helper
def pluralise(count, singular, plural=None):
    """Return the correct singular or plural form based on the count."""
    if count == 1:
        return singular
    return plural if plural else singular + "s"

# 3. select_word() – selects a random word
def select_word():
    '''selects a random word from the word bank'''
    return random.choice(word_bank)

# 4. display() – builds the _ _ _ _ blanks
def display(word):
    '''displays the _ _ _ _ to indicate the length of word'''
    return ["_"] * len(word)

# 5. get_guess() – receives and validates player input
def get_valid_guess(guess, guessed_letters):
    '''Validates the player input to ensure it is a single new letter'''
    if len(guess) != 1:
        print("Only 1 letter at a time please")
        return False
    if not guess.isalpha():
        print("Only letters are valid inputs")
        return False
    if guess in guessed_letters:
        print("Woops, you have already guessed that")
        return False
    return True

def get_guess(guessed_letters):
    '''input for your guesseses'''
    while True:
        guess = input("your guess: ").lower().strip()
        if get_valid_guess(guess, guessed_letters):
            return guess

# 6. update_display() – fills in correct letters
def update_display(word, display_list, guess):
    ''' makes the word into a list of letters log each letters position and identifys the letters then checks if the guess is one of the letters'''
    for index, letter in enumerate(word):
        if letter == guess:
            display_list[index] = guess
    return display_list

# 7. update_wrong_guesses() – tracks incorrect guesses
def update_wrong_guesses(guess, wrong_letters):
    wrong_letters.add(guess)
    '''add wrong guesses to the wrong guesses set'''
    return wrong_letters

# 8. check_win() – no blanks remain
def check_win(display_list):
    '''checks if there are no "_" left in the display'''
    return "_" not in display_list

# 9. check_loss() – 6 incorrect guesses
def check_loss(wrong_letters):
    '''if full hangman showes you lose'''
    max_wrong = len(HANGMAN_PICS) - 1
    return len(wrong_letters) >= max_wrong

# 12. show_results() – prints statistics
def show_results(stats):
    '''prints the game stats'''
    print("\nFinal results for all players:")
    print("------------------------------")
    for name, record in stats.items():
        wins = record['wins']
        losses = record['losses']
        correct = record['correct']
        incorrect = record['incorrect']

        print(
            f"{name}: "
            f"{wins} {pluralise(wins, 'win')}, "
            f"{losses} {pluralise(losses, 'loss', 'losses')}, "
            f"{correct} {pluralise(correct, 'correct guess', 'correct guesses')}, "
            f"{incorrect} {pluralise(incorrect, 'incorrect guess', 'incorrect guesses')}"
        )

# shows the hang man stage 
def show_hangman(wrong_count):
    '''creates the hang man picture'''
    index = min(wrong_count, len(HANGMAN_PICS) - 1)
    print(HANGMAN_PICS[index])

# 10. play_round() – inner loop for one game
def play_round(stats):
    '''core mechanic for running game loop'''
    word = select_word()
    display_list = display(word)
    guessed_letters = set()
    wrong_letters = set()
    correct_guesses = 0
    incorrect_guesses = 0

    while True:
        show_hangman(len(wrong_letters))
        print("Word:", " ".join(display_list))
        print("Incorrect guesses:", " ".join(sorted(wrong_letters)))

        guess = get_guess(guessed_letters)
        guessed_letters.add(guess)

        #point system for the guesses
        if guess in word:
            update_display(word, display_list, guess)
            correct_guesses += 1
        else:
            update_wrong_guesses(guess, wrong_letters)
            incorrect_guesses += 1

        #poit sytem for wins and losses
        if check_win(display_list):
            print("You win! The word was:", word)
            print(f"Correct guesses this round: {correct_guesses}")
            print(f"Incorrect guesses this round: {incorrect_guesses}")

            stats["wins"] += 1
            stats["correct"] += correct_guesses
            stats["incorrect"] += incorrect_guesses
            break

        if check_loss(wrong_letters):
            show_hangman(len(wrong_letters))
            print("You lost! The word was:", word)
            print(f"Correct guesses this round: {correct_guesses}")
            print(f"Incorrect guesses this round: {incorrect_guesses}")

            stats["losses"] += 1
            stats["correct"] += correct_guesses
            stats["incorrect"] += incorrect_guesses
            break

# 11. main() – outer loop for multiple games
def main():
    '''Run the full Hangman game, including multiple players and rounds. Handles player setup, input validation, game loops, and final results.'''
    stats = {}

    while True:
        # Ask for name with validation
        while True:
            name = input("What is your name? ").strip()
            if name:
                name = name.lower()
                break
            print("Invalid name. Please enter at least one character.")

        #creates stat dictonary
        if name not in stats:
            stats[name] = {
                "wins": 0,
                "losses": 0,
                "correct": 0,
                "incorrect": 0
            }

        # Validate play input
        while True:
            play = input(f"Hello {name}, do you want to play Hangman? (y/n): ").lower().strip()
            if play in ("y", "n"):
                break
            print("Invalid input. Please enter 'y' or 'n'.")

        # Play rounds
        while play == "y":
            play_round(stats[name])

            # Validate play again input
            while True:
                play = input("Play again? (y/n): ").lower().strip()
                if play in ("y", "n"):
                    break
                print("Invalid input. Please enter 'y' or 'n'.")

        # Show stats for this player
        wins = stats[name]['wins']
        losses = stats[name]['losses']

        print(f"\n{name}'s stats:")
        print(f"{wins} {pluralise(wins, 'win')}")
        print(f"{losses} {pluralise(losses, 'loss', 'losses')}")

        # Validate another player input
        while True:
            another = input("\nIs there another player? (y/n): ").lower().strip()
            if another in ("y", "n"):
                break
            print("Invalid input. Please enter 'y' or 'n'.")

        if another == "n":
            break

    show_results(stats)

# RUN THE PROGRAM (with try wrapper for marks)
if __name__ == "__main__":
    try:
        main()
    except Exception as e:
        print("\nAn unexpected error occurred:", e)

```
